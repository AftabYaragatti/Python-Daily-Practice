# File Handling Concept

## ==========Text File========== :

### Open() File :

```python
file=open('Day1_Python.txt','x')
```

### Close() File :

```python
file.close()
```

### Create File using x mode

```python
file=open('Day2_Python.txt','x')
file.close()
```

### Read() :

```python
file=open('Day1_Python.txt','r')
content=file.read()
print(content)
file.close()
```

### Write() :

```python
file=open('Day2_Python.txt','w')
content=file.write("Hello Aftab! Welcome to Python Class")
file.close()
```

```python
## To verify whether file is written or not we need to read the file 
file=open('Day2_Python.txt','r')
content=file.read()
print(content)
file.close()
```

### Append :

```python
file=open('Day1_Python.txt','a')
content=file.write('\nWelcome To QSpiders Basavanagudi........')
file.close()
```

### MODES IN FILE HANDLING :

### A. Read Mode (r,r+) :

```python
file=open('Day1_Python.txt','r')
content=file.read()
print(content)
file.close()
```

```python
file=open('Day1_Python.txt','r+')
content=file.read()
print(content)
content2=file.write("\nQSPIDERS | PYSPIDERS | JSPIDERS")
print(content2)
file.close()
```

### B. Write Mode (w,w+) :

```python
file=open('Day2_Python.txt','w')
content=file.write("Hello Aftab! Welcome Again to Python Class")
file.close()
```

```python
## To verify whether file is written or not we need to read the file 
file=open('Day2_Python.txt','r')
content=file.read()
print(content)
file.close()
```

### C. Apend Mode (a,a+) :

```python
file=open('Day1_Python.txt','a')
content=file.write('\nWelcome To QSpiders Basavanagudi........')
file.close()
```
