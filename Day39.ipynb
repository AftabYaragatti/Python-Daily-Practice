# File Handling Concepts :

### 1.Write(w+) Mode :

```python
file=open('File1.txt','w+')
file.write("Hello Welcome to Pyspiders Basavanagudi")
file.close()
```

```python
file=open('File1.txt','w+') # however we are opening the file it will trucate the old content
print(file.read())
file.close() # Nothing will displayed because it trucated(permanently deleted ) old content from the file
# to over come this method we are using seek() method
```

```python
### W+ : It is used to both read and write 
file=open('File1.txt','w+')
file.write("Hello Welcome to Pyspiders Basavanagudi")
file.seek(0)
print(file.read())
file.close()
```

### 2.Read(r+) Mode :

```python
file=open('File2.txt','r+')
file.write("Hello Welcome to Pyspiders Basavanagudi")
file.close()
```

```python
file=open('File2.txt','r+')
print(file.read())
file.close()
### Here seek() method is not required
```

```python
file=open('File2.txt','r+')
print(file.read())
file.write("\nHello Welcome to Jyspiders Basavanagudi")
print(file.read())
file.close()
```

```python
file=open('File2.txt','r+')
print(file.read())
file.close()
```

```python
### R+ : It is used to both read and write and Seek() method is not manadatory in r+ mode.... 
file=open('File1.txt','r+')
file.write("Hello Welcome to Pyspiders Basavanagudi")
file.seek(0)
print(file.read())
file.close()
```

### 3.Append(a+) Mode :

```python
file=open('File1.txt','a+')
file.write("Hello Welcome to Qyspiders | Pyspiders | JSpiders Basavanagudi")
file.close()
```

```python
file=open('File1.txt','a+') # however we are opening the file it will trucate the old content
print(file.read())
file.close() # Nothing will displayed because it trucated(permanently deleted ) old content from the file
# to over come this method we are using seek() method
```

```python
### A+ : It is used to both write and append 
file=open('File1.txt','a+')
file.write("Hello Welcome to Qyspiders | Pyspiders | JSpiders Basavanagudi")
file.seek(0)
print(file.read())
file.close()
```

### File Handling operation using with statements :

#### Syntax with expression as varible:
                Perform operations

### 1.Write(w,w+) mode :

```python
with open('File3.txt','w') as file:
    file.write("Hello Aftab! What's Up!")
# Here no need to close the file, it will automatically close() the file...
```

```python
with open('File3.txt','w+') as file: 
    file.write("Hello Bro! What's Up!")
    file.seek(0)
    print(file.read())
### Note : It over writes the privious content to that many characters only owerwrite to the file rest of all content remain same
```

### 2.Read(r,r+) mode :

```python
with open('File3.txt','r') as file:
    print(file.read())
# Here no need to close the file, it will automatically close() the file...
```

```python
with open('File3.txt','r+') as file: 
    file.seek(0)
    print(file.read())
    file.write("Hello Smith! What's Up!")
    print(file.read())
### Note : It over writes the privious content to that many characters only owerwrite to the file rest of all content remain same
```

```python
with open('File3.txt','r') as file:
    print(file.read())
```

### 3.Append(a,a+) mode :

```python
with open('File4.txt','a') as file:
    file.write("Hello Aftab! What's Up!")
```

### Types of Read() Operation :

### 1.Read() : It reads entire file content

```python
with open('Day1.txt','r') as file:
    print(file.read())
```

```python
with open('Day1.txt','r') as file:
    print(file.read(19)) # It reads only 19 characters from the file including whitespaces
```

### 2.Readline() : It reads entire line file content

```python
with open('Day1.txt','r') as file:
    print(file.readline()) # it reads only first single line from the file
```

```python
with open('Day1.txt','r') as file:
    print(file.readline(4)) ### If we pass parameter then it will take that many characters only
```

### 3.Readlines() : It reads entire one one lines of file content and returns in terms of list

```python
with open('Day1.txt','r') as file:
    print(file.readlines()) # it reads only entire single single line from the file and returns the content of each line in list form
```

```python
with open('Day1.txt','r') as file:
    print(file.readlines(5)) # it reads only 5 characters from first line, if first line is less than 5 characters than it takes entire 2nd line 
```

```python
with open('Day1.txt','r') as file:
    print(file.readlines(17)) # it reads only 17 characters from first line, if first line is less than 16 characters than it takes enitre 2nd line
                              # and then from it will print upto 16 characters from both the lines, if again it less than 17 characters then it takes 
                              # 3rd entir line and wil print
```
