# OPERATORS IN PYTHON 

## 1. Arithmetic Operator :

```python
# + (Plus) :
a=10
b=20
print(a+b) 
x='Hello'
y='Python'
print(x+y) 
p='20'
q=30
print(int(p)+q) 
print(p+str(q))  
print(12+34.2) 
```

```python
# - (minus) :
a=20
b=12
print(a-b)
print(12-239.30)
x='20'
y='34'
# print(x-y) 
print(int(x)-int(y))
```

```python
# * (multiplication) :
# Replication done by using *
a=10
b='Python'
print(a*b)
# Multiplication done by *
x=2
y=7
print(x*y)
```

```python
# / (Division) : It gives quetiont
a=12
b=23
print(a/b)
x=12.3
y=2.3
print(x/y)
```

```python
# % (Modulus) : It gives reminder
a=29
b=3
print(a%b)
x=93
y=6.4
print(x%y)
```

```python
# // (Floor Division) : It will gives the decimal/float value
a=13
b=2
print(a//b)
```

```python
# ** (Exponentiation) :
a=2
b=3
print(a**b)
```
