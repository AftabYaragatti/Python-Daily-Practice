```python
def sum(num1,num2):
    print(num1+num2)
sum(12,19)
sum(83,92)
```

```python
def even_odd(num):
    if num%2==0:
        print("EVEN")
    else:
        print("ODD")
even_odd(10)
```

```python
def prime(num):
    start=1
    count=0
    while start<=num:
        if num%start==0:
            count+=1
        start+=1
    if count==2:
        print("Prime")
    else:
        print('Not Prime')
prime(7)
```
