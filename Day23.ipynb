### 7. Reverse only the digits present in a string.

```python
st='987Hello'
res=''
for ch in st:
    i=0
    if ch[i]>='0' and ch[i]<='9':
        res=ch[i]+res
    elif (ch[i]>='A' and ch[i]<='Z') or (ch[i]>='a' and ch[i]<='z'):
        res=res+ch[i]
    i+=1
print(res)
```

### 8. Replace duplicate characters with '*'.

```python
string = 'programming'

result = ""

for i in range(len(string)):
    for j in range(i):
        if string[i] == string[j]:
            result += "*"
            break
    else:
        result += string[i]

print(result)
```

### 11. Find the longest palindrome in a string.

```python
string = 'bababul'
longest = ""
for i in range(len(string)):
    for j in range(i + 1, len(string) + 1):
        sub = string[i:j]
        if sub == sub[::-1]:  
            if len(sub) > len(longest):
                longest = sub

print(longest)
```

### 12. Find the smallest palindrome in a string.

### 13. Find the smallest and largest words in a string.

```python
string = input("Enter a sentence: ")
words = string.split()
smallest = words[0]
largest = words[0]
for word in words:
    if len(word) < len(smallest):
        smallest = word
    if len(word) > len(largest):
        largest = word
print("Smallest word:", smallest)
print("Largest word:", largest)
```

### 24. Convert a string to its equivalent integer representation.

```python
text = '1234'

num = 0

for ch in text:
    digit = ord(ch) - ord('0')
    num = num * 10 + digit

print("Integer value:", num)
```
