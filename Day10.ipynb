# OPERATORS IN PYTHON 

## 4. Assignment  Operator 

```python
# Example : +=, -=, *=, /= ,//=, **=
a=10
b=a
#This is the basic operator
b=a+b
print(b)
#Using Assignment operator we can do as follows :
x=10
y=x
y+=x
print(y)
```

```python
# Basic Examples
a=10
b=a
print(b)
b+=a
print(b)
b-=a
print(b)
b*=a
print(b)
b/=a
print(b)
b//=a
print(b)
b**=b
print(b)
```

```python
# Basic left shift and right shift
a=10
b=a
b<<=a
print(b)
b>>=a
print(b)
```

## 5. Membership Operator :

```python
#Example :
x=2
y=8
list=[1,2,3,4,5,6]
if (x not in list):
    print("x is NOT present in given list")
else:
    print("x is present in given list")
    
if (y in list):
    print("y is present in given list")
else:
    print("y is NOT present in given list")
```

## 6. Identity Operator :

```python
# Example : 
a=10
b=32
c=a
print(a is b) 
print(a is c) 
```

```python
#Example on Tuple : 
t1=(1,2,3) 
t2=(4,5,2)
t3=t2
print(t1 is t2)
print(t2 is t3)
print(t1 is not t3)
```
