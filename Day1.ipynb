```python
# Printing with different kinds of data
name='Aftab'
age=21
marks=[89,85,93,95,80]
print(name,age,marks)
print('Name :',name,'\nAge :',age,'\nMarks :',marks)
```

```python
# Formating the String in print statement
name='Aftab'
age=21
marks=[89,85,93,95,80]
print(f'{name} , {age} , {marks}')
print(f'Name : {name} \nAge : {age} \nMarks : {marks}')
```

# Variables :

```python
# Initialize the variable and print the value
name='Aftab'
age=21
print(name,' | ',age)
```

```python
# ReInitialize the variable and print the value
name='Aftab'
age=21
print(name,' | ',age)
age=25
print(name,' | ',age)
```

```python
# Assign Multiple Values to the Multiple Variables
a,b,c=12,34,64
print(a,b,c)
```

```python
# Assign Multiple Values to the Single Variables using *
name,age,*marks='Aftab',21,45,56,76,54,45
print(name,age,marks)
```

```python
# Assign Single Value to the Multiple Variables
a=b=c=10
print(a,b,c)
```

```python
# Assign values in list and Assign Single Value to the Multiple Variables
names=['Smith','Jhon','Pretty']
a,b,c=names
print(a,b,c)
```
