## 1. Positional Arguemnets :

```python
## WAP print the sum of two numbers
def add(num1,num2):
    print(num1+num2)
add(10,20)
```

```python
## WAP to find multiplication of 3 numbers
def mul(a,b,c):
    print(a*b*c)
mul(10,20,30)
```

## 2. Keyword Arguements :

```python
## WAP to display Student Name,Age and Subject
def details(name,age,sub):
    print(f'Student Name : {name}\nStudent Age : {age}\nStudent Subject : {sub}')
details(name='Aftab',sub="Python",age=21)
```

```python
## WAP to find multiplication of 3 numbers
def mul(a,b,c):
    print(a)
    print(b)
    print(c)
    print(a*b*c)
mul(a=10,c=20,b=30)
```

```python
def details(name,age,sal):
    print(f'Employee Name : {name}\nEmployee Age : {age}\nEmployee Salary : {sal}')
details(name='Aftab',age=21,sal=20000)
```

## 3. Default Arguement :

```python
## Find the power of the given number
def pow(num,pow=2):
    print(num**pow)
pow(4,3)
```

```python
## Display The country name
def country(name='India'):
    print(name)
country('USA')
```

```python
def details(name,age,sal=20000):
    print(f'Employee Name : {name}\nEmployee Age : {age}\nEmployee Salary : {sal}')
details('Aftab',21)
```

## 4. Varibale Length Arguements :

### a. *args :

```python
## WAP to print the sum of multiple numbers
def add(*nums):
    sum=0
    for i in nums:
        sum+=i
    print(sum)
add(1,2,3,4,5)
```

```python
def details(*subjects):
    print(f'First Subject : {subjects[0]}\nSecond Subject : {subjects[1]}\nThird Subject : {subjects[2]}\nFourth Subject : {subjects[3]}')
details('Python','Java','SQL','Web')
```

```python
## WAP to print the product of multiple numbers
def prod(*nums):
    p=1
    for i in nums:
        p*=i
    print(p)
prod(1,2,3,4,5)
```

### b. **kwargs :

```python
## WAP to dispaly marks of given subjects
def display(**marks):
    print(f'Python : {marks['m1']}\nJava : {marks['m2']}\nSQL : {marks['m3']}\nPowerBI : {marks['m4']}')
display(m1=90,m3=98,m2=89,m4=91)
```

```python
def details(**subjects):
    print(f'First Subject : {subjects['s1']}\nSecond Subject : {subjects['s2']}\nThird Subject : {subjects['s3']}\nFourth Subject : {subjects['s4']}')
details(s1='Python',s2='Java',s3='SQL',s4='Web')
```
