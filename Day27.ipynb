# SUBARRAY/ SUBSTRING

```python
st='Python'
i=0
while i<=len(st):
    j=i+1
    while j<=len(st):
        substring=st[i:j]
        print(substring)
        j+=1
    print()
    i+=1
```

```python
l=[1,2,3,4,5,6,7]
i=0
while i<=len(l):
    j=i+1
    while j<=len(l):
        substring=l[i:j]
        print(substring)
        j+=1
    print()
    i+=1
```

```python
# WAP to print the substrings whose lenght is 3 from a given string
st='Python'
i=0
while i<=len(st):
    j=i+1
    while j<=len(st):
        substring=st[i:j]
        if len(substring)==3:
            print(substring)
        j+=1
    i+=1
```

```python
# WAP to print the substrings that is palindrome from a given string
st='BBeBBeTsTsTRsKJLSAJd'
i=0
while i<=len(st):
    j=i+1
    while j<=len(st):
        substring=st[i:j]
        if substring==substring[::-1] and len(substring)>1:
            print(substring)
        j+=1
    i+=1
```
