```python
## WAP to print the table of a given number
def number():
    num=10
    def table():
        i=1
        while i<=10:
            print(f'{num} X {i} = {num*i}')
            i+=1
    table()
number()
```

```python
## WAP to find the length of the given string
def str():
    s='Hello Welcome to QSpiders'
    def s_len():
        count=0
        for ch in s:
            count+=1
        print(count)
    s_len()
str()
```

```python
## WAP to reverse the number
def number():
    num=1234
    def reverse():
        nonlocal num
        rev=0
        while num!=0:
            d=num%10
            rev=rev*10+d
            num//=10
        print(rev)
    reverse()
number()
```

```python
## WAP to check the given number is Palindrome or not
def number():
    num=13231
    num1=num
    def reverse():
        nonlocal num,num1
        rev=0
        while num!=0:
            d=num%10
            rev=rev*10+d
            num//=10
        if num1==rev:
            print('Palindrome')
        else:
            print('Not A Palindrome')
    reverse()
number()
```

## NUMBER PROGRAM USING NESTED FUNCTIONS

## 1. WAP to print the sum of the given digits

```python
def digits():
    num=234823
    def add():
        sum=0
        nonlocal num
        while num>0:
            d=num%10
            sum+=d
            num//=10
        print(f'Sum of the given Digits {num} is : {sum}')
    add()
digits()
```

## 2. WAP to check given number is even or odd, if it is even do sum of the digits

```python
def even_odd():
    num=78
    if num%2==0:
        print(f'{num} is Even Number')
    else:
        print(f'{num} is odd Number')
    def add():
        nonlocal num
        sum=0
        if num%2==0:
            while num>0:
                d=num%10
                sum+=d
                num//=10
            print(f'Sum of the Even Digits is {sum}')
    add()
even_odd()
```

## 3. WAP to chcek given number is NEON NUMBER 

```python
def number():
    num=9
    print(f'Number is {num}')
    def neon():
        nonlocal num
        sum=0
        sqr=num*num
        while sqr!=0:
            d=sqr%10
            sum+=d
            sqr//=10
        if num==sum:
            print(f'The Number {num} is Neon Number')
        else:
            print(f'The Number {num} not a Neon Number')
    neon()
number()
```

## 4. WAP to check given number is Strong Number or not

```python
def number():
    num=145
    num1=num
    print(f'Number is {num}')
    def strong():
        nonlocal num,num1
        sum=0
        while num!=0:
            d=num%10
            fact=1
            start=1
            while start<=d:
                fact*=start
                start+=1
            num//=10
            sum+=fact
        if sum==num1:
            print(f'Number {num1} is an STRONG NUMBER')
        else:
            print(f'Number {num1} is not a SRONG NUMBER')
    strong()
number()
```

## 5. WAP to check given number is PERFECT NUMBER or not

```python
def number():
    num=6
    print(f'Number is {num}')
    def perfect():
        nonlocal num
        start=1
        sum=0
        while start<num:
            if num%start==0:
                sum+=start
            start+=1
        if sum==num:
            print(f'Number {num} is Perfect Number')
        else:
            print(f'Number {num} is Not Perfect Number')
    perfect()
number()
```

## 6. WAP to check Given Number is ARMSTRONG NUMBER or not

```python
def number():
    num=153
    num1=num
    temp=num
    print(f'Number is {num}')
    def count():
        nonlocal num
        count=0
        while num>0:
            count+=1
            num//=10
        def armstrong():
            nonlocal num1,count,temp
            sum=0
            while num1>0:
                d=num1%10
                p=d**count
                sum+=p
                num1//=10
            if sum==temp:
                print(f'Number {temp} is ARMSTRONG NUMBER')
            else:
                print(f' number {temp} is NOT ARMSTRONG NUMBER')
        armstrong()
    count()
number()
```

## 7. WAP to check given number is PRIME NUMBER or NOT

```python
def number():
    num=17
    print(f'Number is {num}')
    def prime():
        nonlocal num
        start=1
        count=0
        while start<=num:
            if num%start==0:
                count+=1
            start+=1
        if count==2:
            print(f'Number {num} is PRIME NUMBER')
        else:
            print(f'Number {num} is NOT PRIME NUMBER')
    prime()
number()
```
