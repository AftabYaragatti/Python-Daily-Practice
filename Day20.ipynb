#### 9. WAP to count the total numbers from a given string

```python
st=input()
count=0
for ch in st:
    if ch>='0' and ch<='9':
        count+=1
print(count)
```

#### 14. WAP to count the total number of Alphabets,Character,Uppercase,Lowercase,Digits and Symbols

```python
st=input()
al=0
uc=0
lc=0
d=0
s=0
c=0
for ch in st:
    c+=1
    if (ch>='A' and ch<='Z') or (ch>='a' and ch<='z'):
        al+=1
    if (ch>='A' and ch<='Z'):
        uc+=1
    if (ch>='a' and ch<='z'):
        lc+=1
    if (ch>='0' and ch<='9'):
        d+=1
    else:
        if not((ch>='A' and ch<='Z') or (ch>='a' and ch<='z') or (ch>='0' and ch<='9')):
            s+=1
print(f'Characters : {c}\n Alphabets : {al}\n UpperCase : {uc}\n Lowercase : {lc}\n Digits : {d}\n Symbols : {s}')
```

### OR

```python
st=input()
al=0
uc=0
lc=0
d=0
s=0
c=0
for ch in st:
    c+=1
    if (ch>='A' and ch<='Z') or (ch>='a' and ch<='z'):
        al+=1
        if (ch>='A' and ch<='Z'):
            uc+=1
        else:
            lc+=1
    elif (ch>='0' and ch<='9'):
        d+=1
    else:
        s+=1
print(f'Characters : {c}\n Alphabets : {al}\n UpperCase : {uc}\n Lowercase : {lc}\n Digits : {d}\n Symbols : {s}')
```

#### 14. WAP to count the total number of Alphabets,Character,Uppercase,Lowercase,Digits and Symbols and store all values in dictionary

```python
st=input()
al=0
uc=0
lc=0
d=0
s=0
c=0
for ch in st:
    c+=1
    if (ch>='A' and ch<='Z') or (ch>='a' and ch<='z'):
        al+=1
        if (ch>='A' and ch<='Z'):
            uc+=1
        else:
            lc+=1
    elif (ch>='0' and ch<='9'):
        d+=1
    else:
        s+=1
dict={}
dict['Character']=c
dict['Alphabets']=al
dict['UpperCase']=uc
dict['LowerCase']=lc
dict['Digits']=d
dict['Symbols']=s
print(dict)
```

#### 15.WAP to reverse the string
