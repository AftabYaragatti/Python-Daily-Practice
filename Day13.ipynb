# While Loop

```python
# While Loop Example :
num=int(input('Enter the Number You want ro repeat :'))
count=0
while count<num:
    count+=1
    print('Hello World')
```

```python
# Example 1 withoud break :
i=0
while i<5:
    if i%2==0:
        print(f'{i} is Even Number')
    i+=1
else:
    print('Loop Completed withoud break')
```

```python
# Example 2 withoud break :
i=0
while i<5:
    if i%2==0:
        print(f'{i} is Even Number')
        break
else:
    print('Loop Completed withoud break')
```

```python
num=int(input()) 
start=1 
while start<=num: 
    print(start) 
    start+=1 
```

```python
start=5 
while start>0: 
    print(start)
    start-=1
```
