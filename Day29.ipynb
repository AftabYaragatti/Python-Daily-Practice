```python
def is_prime(num):
    start=1
    count=0
    while start<=num:
        if num%start==0:
            count+=1
        start+=1
    if count==2:
        print('Prime Number')
    else:
        print('Not a Prime Number')
is_prime(7)
is_prime(6)
```

```python
### SPY NUMBER PROGRAM
def spy_number(num):
    sum=0
    prod=1
    while num!=0:
        d=num%10
        sum+=d
        prod*=d
        num//=10
    if sum==prod:
        print('Spy Number')
    else:
        print('Not a Spy Number')
spy_number(123)
spy_number(342)
```

```python
### Neon Number Program
def neon_number(num):
    num1=num
    sqr=num*num
    sum=0
    while sqr!=0:
        d=sqr%10
        sum+=d
        sqr//=10
    if sum==num1:
        print('Neon Number')
    else:
        print('Not a Neon Number')
neon_number(9)
neon_number(12)
```

```python
### Perfect Number Program:
def perfect_number(num):
    start=1
    sum=0
    while start<num:
        if num%start==0:
            sum+=start
        start+=1
    if sum==num:
        print('Perfect Number')
    else:
        print('Not a Perfect Number')
perfect_number(6)
perfect_number(153)
```

```python
### Strong Number Program :
def strong_number(num):
    num1=num
    sum=0
    while num!=0:
        d=num%10
        fact=1
        start=1
        while start<=d:
            fact*=start
            start+=1
        num//=10
        sum+=fact
    if sum==num1:
        print('Strong Number')
    else:
        print('Not a Strong Number')
strong_number(145)
```

```python
### Armstrong Number Program :
def armstrong_number(num):
    num1=num
    num2=num
    count=0
    while num1!=0:
        count+=1
        num1//=10
    sum=0
    while num2!=0:
        d=num2%10
        pow=d**count
        num2//=10
        sum+=pow
    if sum==num:
        print('Armstrong Number')
    else:
        print('Not Armstrong Number')
armstrong_number(153)
```

```python
### Factorial of a Number :
def fact_num(num):
    fact=1
    start=1
    while start<=num:
        fact*=start
        start+=1
    print(fact)
fact_num(5)
```
