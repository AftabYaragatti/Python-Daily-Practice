# MULTI VALUE DATA TYPES IN PYTHON 

## String Data type with Strings CRUD Operations 

```python
# Basics of Strings :
s="Hello World"
print(s)
print(type(s),len(s))
```

```python
s='Hello "Python Developer"' 
print(s)
```

```python
#If we want to write the string in multiple lines we uses thriple quotes '''msg'''
s='''String is an collection of sequence of character enclosed between single quotes or a double quotes
String is an immutable data type and supports indexing'''
print(s)
```

```python
# Indexing in String
s="Python"
print(s[0])
print(s[-1])
print(s[3]) 
```

```python
# Basic CRUD Operations on String

#Create a string or define string :
s="Python Developer"
s1='' #empty string
print(s,s1)
```

```python
#Read string :

#Read Entire String
s="Python Developer"
print(s) 

#Read Specific character of String (single character)
print(s[3]) 
print(s[-5]) 
```

```python
# Update is not supporting in String as it is immutable data type
```

```python
# Delete String :
s="Python Developer"
print(s)
del s
print(s) 
```

## List  Data type with CRUD Operations :::

```python
lst=[1,2,3,4,'Hello',23.9,'Python']
print(lst)
print(type(lst))
```

```python
#Adding two lists in single list
lst1=[1,2,'A','B']
lst2=[3,4,'C','D']
res=[lst1,lst2]
print(res)
```

```python
names=['Alice','Bob','Jhon','Robart']
print(names)
print(len(names))
```

```python
#To access the elements of list we uses indexing 
names=['Alice','Bob','Jhon','Robart']
print(names[2])
print(names[-1])
print(names[-1][2])
```

```python
#CRUD Operations of List ::

# Create List ::
# 1. Create Empty List
list=[]
print(list)

# 2. Create list/Define list
lst1=[1,2,3,"Hello"]
print(lst1)
```

```python
# Read List Elements ::
# 1. Read entire List 
lst=[1,2,3,'A',23.4,'Hello']
print(lst)

# 2. Read specific elements or object of list
print(lst[2])
print(lst[5][2])
```

```python
# Update list elements
lst=[1,2,3,'A',23.4,'Hello']
print(lst)
lst[1]='Python'
print(lst)
```

```python
# Delete list element

# 1. Delete single object on list items
lst=[1,2,3,'A',23.4,'Hello']
print(lst)
del lst[4]
print(lst)

# 2. Delete entire list
print(lst)
del lst
print(lst)
```

```python
# Nested List ::
lst=[1,2,3,4,[3,6,7],9]
print(lst)
print(lst[-2][2])
```
