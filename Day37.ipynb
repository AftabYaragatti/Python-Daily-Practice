# Dictionary Built-in-Methods :

### 1.get():

```python
d={'eno':2903,'ename':'Aftab','sal':'98000'}
print(d.get('eno'))
print(d.get('ename'))
print(d.get('sal'))
```

### 2.update():

```python
d={'eno':2903,'ename':'Aftab','sal':'98000'}
d.update({'job':'Python Developer'})
print(d)
```

```python
d={'eno':2903,'ename':'Aftab','sal':'98000'}
d.update({'job':'Python Developer','age':25})
print(d)
```

```python
d={'eno':2903,'ename':'Aftab','sal':'98000'}
d.update({'job':'Full Stack Developer'})
print(d)
```

### 3.items():

```python
d={'eno':2903,'ename':'Aftab','sal':'98000'}
print(d.items())
```

### 4.keys():

```python
d={'eno':2903,'ename':'Aftab','sal':'98000'}
print(d.keys())
```

### 5.values():

```python
d={'eno':2903,'ename':'Aftab','sal':'98000'}
print(d.values())
```

### 6.pop():

```python
d={'eno':2903,'ename':'Aftab','sal':'98000'}
d.pop('eno')
print(d)
```

### 7.popitem():

```python
d={'eno':2903,'ename':'Aftab','sal':'98000'}
d.popitem()
print(d)
d.popitem()
print(d)
```

### 8.clear():

```python
d={'eno':2903,'ename':'Aftab','sal':'98000'}
d.clear()
print(d)
```

### Basic Program :

```python
### WAP to print which starts with emp using built in methods
data={'empno':2903,'empname':'Aftab','empsal':'98000','empage':25,'empjob':'Software Developer',
      'stdno':1209,'stdname':'King','stdsal':'8000','stdage':21,'stdjob':'Software Developer'}
for key in data:
    if key.startswith('emp'):
        print(key)
```

# COPY OPERATIONS IN PYTHON:

### 1.General Copy:

```python
# creates new object with new memory allocation for both the object when we manually storing data without reference
l1=[10,20,30]
l2=[10,20,30]
print(id(l1))
print(id(l2))
```

```python
# creates new object with same memory allocation for both the object when we manually storing data without reference
l1=[10,20,30]
l2=l1
print(id(l1))
print(id(l2))
```

```python
# if we modify the one object(copied object) it will also modify the original list
l1=[10,20,30]
l2=l1
l2[2]=40
print(l1)
print(l2)
```

### 2.Shallow Copy:

```python
# copy the object values or data
import copy
l1=[10,20,30,[100,200,300]]
l2=copy.copy(l1)
print(l1)
print(l2)
```

```python
# copy the object values or data and modify it 
# if we modify the one object(copied object) in shallow copy it will also modify the original list
import copy
l1=[10,20,30,[100,200,300]]
l2=copy.copy(l1)
l2[3][1]=500
print(l1)
print(l2)
```

### Deep Copy:

```python
# copy the object values or data
import copy
l1=[10,20,30,[100,200,300]]
l2=copy.deepcopy(l1)
print(l1)
print(l2)
```

```python
# copy the object values or data and modify it 
# if we modify the one object(copied object) in deep copy it will not modify the original list
import copy
l1=[10,20,30,[100,200,300]]
l2=copy.deepcopy(l1)
l2[3][1]=500
print(l1)
print(l2)
```
