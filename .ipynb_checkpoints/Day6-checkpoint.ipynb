# MULTI VALUE DATA TYPES IN PYTHON WITH CRUD OPERATONS 

## CRUD OPERATIONS ON DICTIONARY

```python
# CRUD Operations of Dictionary:::

#Create Dictionary :
# 1. Create empty dictionary 
d={}
d1=dict({})
print(d,d1)
# 2. Create/define dictionary using dict() 
d2=dict({1:'Hello',2:300})
print(d2)
```

```python
#Read Dictionary :
# 1. Read entire dictionary 
d={1:'A',2:'B',3:'C'}
print(d)
# 2. Read only specific key values 
print(d[1])
```

```python
#Update Dictionary :
# 1. Update entire dictionary 
d={1:'A',2:'B',3:'C'}
print(d)
d1=dict({4:'D',5:'E',6:'F'})
print(d1)
# 2. Update specific key values 
d1[7]='G'
print(d1)
```

```python
#Delete Dictionary :
# 1. Delete entire dictionary 
d1={1:'A',2:'B',3:'C'}
print(d1)
del d1[3]
print(d1)
# 2. Update specific key values 
d2={1:'A',2:'B',3:'C'}
print(d2)
del d2
print(d2)
```

## Range Data type :::

```python
l=range(5) 
print(l) 
print(list(l)) 
```

```python
l=range(1,6) 
print(list(l))
```

```python
l=range(5,1)
print(list(l)) 
```

```python
l=range(1,9,1) 
print(list(l)) 
print(tuple(l))
print(set(l))
print(frozenset(l))
```

```python
l=range(9,1,-1) 
print(list(l))
l1=range(20,10,-3) 
print(list(l1))
l2=range(1,10,3)
print(list(l2))
```
