# OPERATORS IN PYTHON 

## 2. Relational Operator [Comparison Operators] : 

```python
# Example of Comparison Operators :
a=10
b=20
# Equal to
print(a==b) 
# Not Equal to
print(a!=b) 

#another exapmle :
x=20
y='20'
print(x==y) 
print(x==int(y))
print(str(x)==y) 
```

```python
# Example of Relational Operators :
a=10
b=32
print(a<b) 
print(a>b) 
print(a<=b) 
print(a>=b) 
```

## 3. Logical Operator :

```python
# Example :

#Logical AND
a=True 
b=False
print(a and b)  

x=True #1
y=True #1
print(x and y) 
print(a and b)  

#Logical OR
a1=True
b1=False
print(a1 or b1)

x1=False
y1=False
print(x1 or y1)

#Logical NOT
a2=True
b2=False
print(not a2) 
print(not b2) 
```
