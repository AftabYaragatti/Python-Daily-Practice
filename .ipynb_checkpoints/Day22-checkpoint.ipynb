## FOR LOOPING STATEMENS (Two Pointers)

#### 21. Merging two string by in order where input should be in order or sorted form
#### I/P : S1:axz, S2:cdy
#### O/P : acdxyz

```python
st1=input()
st2=input()
res=''
i=0
j=0
while i<len(st1) or j<len(st2):
    if i<len(st1) and j<len(st2):
        if st1[i]<st2[j]:
            res+=st1[i]
            i+=1
        else:
            res+=st2[j]
            j+=1
    elif i<len(st1):
        res+=st1[i]
        i+=1
    else:
        res+=st2[j]
        j+=1
print(res)
```

```python
# OR
st1=input()
st2=input()
res=''
i=0
j=0
while i<len(st1) or j<len(st2):
    if i<len(st1) and j<len(st2):
        if ord(st1[i])<ord(st2[j]):
            res+=st1[i]
            i+=1
        else:
            res+=st2[j]
            j+=1
    elif i<len(st1):
        res+=st1[i]
        i+=1
    else:
        res+=st2[j]
        j+=1
print(res)
```

```python
#IF the input not in Sorted order then we do as follows
st1='pweslg'
st2='cmsnhk'
res1=list(st1)
res2=list(st2)
n1=len(res1)
n2=len(res2)
for i in range(n1):
    for j in range(0,n1-i-1):
        if res1[j]>res1[j+1]:
            temp=res1[j]
            res1[j]=res1[j+1]
            res1[j+1]=temp
for i in range(n2):
    for j in range(0,n2-i-1):
        if res2[j]>res2[j+1]:
            temp=res2[j]
            res2[j]=res2[j+1]
            res2[j+1]=temp
print(res1,':',res2)
res=''
i=0
j=0
while i<len(res1) or j<len(res2):
    if i<len(res1) and j<len(res2):
        if res1[i]<res2[j]:
            res+=res1[i]
            i+=1
        else:
            res+=res2[j]
            j+=1
    elif i<len(res1):
        res+=res1[i]
        i+=1
    else:
        res+=res2[j]
        j+=1
print(res)
```

#### Sort the strings

```python
st1='pweslg'
st2='cmsnhk'
res1=list(st1)
res2=list(st2)
n1=len(res1)
n2=len(res2)
for i in range(n1):
    for j in range(0,n1-i-1):
        if res1[j]>res1[j+1]:
            temp=res1[j]
            res1[j]=res1[j+1]
            res1[j+1]=temp
for i in range(n2):
    for j in range(0,n2-i-1):
        if res2[j]>res2[j+1]:
            temp=res2[j]
            res2[j]=res2[j+1]
            res2[j+1]=temp
res=res1+res2
n=len(res)
for i in range(n):
    for j in range(0,n-i-1):
        if res[j]>res[j+1]:
            temp=res[j]
            res[j]=res[j+1]
            res[j+1]=temp
print(res)
```
