# DATA TYPES IN PYTHON :

## 1. Single Value Datatype 
### i)int ii)float iii)complex iv)boolean v)none

```python
# i. int data type ::
a=100
print(a,type(a))
```

```python
# ii. float data type ::
b=13.30
print(b,type(b))
```

```python
# iii. complex data type ::
c=3+2j
print(c,type(c))
```

```python
# iv. boolean data type ::
d=True
print(d,type(d))
```

```python
# v. None Data type ::
e=None
print(e,type(e))
```

## 2. Multi Value Datatype : 
### i.string  ii.list iii.tuple iv.range v.set vi.frozenset vii. dictionary

```python
# i. string data type :::
s='Hello'
print(s,type(s))
```

```python
# ii. list datatype :::
l=[10,20,30,40,50]
print(l,type(l))
```

```python
# iii. tuple datatype :::
t=(10,20,30,40,50)
print(t,type(t))
```

```python
# iv. set datatype :::
st={10,20,30,40,50,20}
print(st,type(st))
```

```python
# v. frozenset datatype :::
f=frozenset({10,20,30,40,50,20})
print(f,type(f))
```

```python
# vi. dictionary datatype :::
dict={'A':101,'B':102}
print(dict,type(dict))
```

```python
# vii. range datatype :::
r=range(1,20)
print(r,type(r))
```
