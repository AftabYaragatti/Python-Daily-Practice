# String Built-in-Methods 

### 1. Capitalize():

```python
text='weLCOMe to pyThoN clAsS'
print(text.capitalize())
```

### 2.Upper():

```python
text='weLCOMe to pyThoN clAsS'
print(text.upper())
```

### 3.Lower():

```python
text='weLCOMe to pyThoN clAsS'
print(text.lower())
```

### 4.Casefold():

```python
text='weLCOMe to pyThoN clAsS'
print(text.casefold())
```

### 5.Swapcase():

```python
text='weLCOMe to pyThoN clAsS'
print(text.swapcase())
```

### 6.Count():

```python
text='welcome to python class'
print(text.count('o'))
print(text.count('o',7,18))
```

### 7.Index():

```python
text='welcome to python class'
print(text.index('p'))
```

### 8.Find():

```python
text='welcome to python class'
print(text.find('m'))
```

### 9.Rindex():

```python
text='welcome to python class'
print(text.rindex('t'))
```

### 10.Replce():

```python
text='welcome to python class'
print(text.replace('o','*'))
```

### 11.Startswith():

```python
text='welcome to python class'
print(text.startswith('w'))
```

### 12.Endswith():

```python
text='welcome to python class'
print(text.endswith('s'))
```

### 13.Split():

```python
text='welcome to python class'
print(text.split())
```

### 14.Join():

```python
text='welcome to python class'
res="*".join(text)
print(res)
```

### 15.Isalpha():

```python
text='welcome'
print(text.isalpha())
```

### 16.Isalnum():

```python
text='welcome1234'
print(text.isalnum())
```

### 17.Islower():

```python
text='welcome'
print(text.islower())
```

### 18.Islower():

```python
text='WELCOME'
print(text.isupper())
```

### 19.Isdigit():

```python
text='12345'
print(text.isdigit())
```

### 20.Isidentifier():

```python
text='welcome_123'
print(text.isidentifier())
```

### 21.Zfill():

```python
text='457'
print(text.zfill(8))
```

### Programs on String :

```python
## WAP to even index characters of given string
text='welcome to python class'
print(text[0:len(text)-1:2])
```

```python
## WAP to print every alternative characters in reverse order
text='welcome to python class'
print(text[::-2])
```

```python
## WAP to print even index characters in reverse order
text='welcome to python class'
print(text[-2::-2])
```

```python
## WAP to print extension of file name
text='python.txt'
res=text.split('.')
print(res[-1])
```

```python
## WAP to print only file name
text='python.txt'
res=text.split('.')
print(res[0])
```

```python
## WAP to print protocol of url
text='https://Qspiders.com'
res=text.split('://')
print(res[0])
```
