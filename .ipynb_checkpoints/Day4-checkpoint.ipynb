# MULTI VALUE DATA TYPES IN PYTHON WITH CRUD OPERATONS

## CRUD Operations of List ::

```python
#CRUD Operations of List ::

# Create List ::
# 1. Create Empty List
list=[]
print(list)

# 2. Create list/Define list
lst1=[1,2,3,"Hello"]
print(lst1)
```

```python
# Read List Elements ::
# 1. Read entire List 
lst=[1,2,3,'A',23.4,'Hello']
print(lst)

# 2. Read specific elements or object of list
print(lst[2])
print(lst[5][2])
```

```python
# Update list elements
lst=[1,2,3,'A',23.4,'Hello']
print(lst)
lst[1]='Python'
print(lst)
```

```python
# Delete list element

# 1. Delete single object on list items
lst=[1,2,3,'A',23.4,'Hello']
print(lst)
del lst[4]
print(lst)

# 2. Delete entire list
print(lst)
del lst
print(lst)
```

```python
# Nested List ::
lst=[1,2,3,4,[3,6,7],9]
print(lst)
print(lst[-2][2])
```

## Tuple Data type with CRUD Operations :::

```python
# Basics of Tuple 
t=(1,2,3)
t1=4,5,6
print(t,t1)
```

```python
#Nested Tuple
t=(1,2,3,(5,6),9)
print(t)
print([3][-1])
t1=320, 
print(t1)
```

```python
# CRUD Operations of Tuple ::

# Create Tuple : 
# 1. Create/define Tuple
t=tuple((1,2,3,4))
print(t)
# 2. Create Empty Tuple
t1=()
print(t1)
```

```python
# Read element of Tuple :

# 1. Read entire tuple elements 
t1=(1,"Hello",23.54, 23)
print(t1)
# 2. Read Specific object in tuple
print(t1[1])
print(t1[1][-1])
```

```python
# Update elements of tuple : Tuple is not support to update the elements as it is an immutable Object

#Delete Element of Tuple :
# 1. Delete entire elements in tuple :
t1=(1,2,3,4,5,6)
print(t1)
del t1
print(t1)
# 2. Deleting specific element is not supporting in tuple
```
