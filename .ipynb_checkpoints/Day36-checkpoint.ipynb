# Tuple Built-in-methods:

### 1.count():

```python
t1=(100,20,30,40,50,60,70,80,90,100)
print(t1.count(100))
```

### 2.index():

```python
t1=(100,20,30,40,50,60,70,80,90,100)
print(t1.index(50))
```

```python
### WAP to print the 3rd occurance of the number 20
t1=(100,20,30,40,20,50,60,70,80,20,90,100,20)
value=20
count=0
for i,item in enumerate(t1):
    if item==value:
        count+=1
        if count==3:
            print(i)
            break
```

### enumerate():

```python
st='Python'
for i,ch in enumerate(st):
    print(i,ch)
```

```python
st='Python'
for i,ch in enumerate(st,start=5):
    print(i,ch)
```

### Set built-in-methods:

### 1.add():

```python
s={1,2,3,4,5,2,3,4,5,4,13,10,8,9,3,2}
s.add(100)
print(s)
```

### 2.remove():

```python
s={1,2,3,4,5,2,3,4,5,4,13,10,8,9,3,2}
s.remove(3)
print(s)
```

### 3.update():

```python
s={1,2,3,4,5,2,3,4,5,4,13,10,8,9,3,2}
s.update({100,200,300})
print(s)
```

### 4.pop():

```python
s={1,2,3,4,5,2,3,4,5,4,13,10,8,9,3,2}
s.pop()
print(s)
```

### 5.clear():

```python
s={1,2,3,4,5,2,3,4,5,4,13,10,8,9,3,2}
s.clear()
print(s)
```

### 6.union():

```python
s={1,2,3,4,5}
res=s.union({100,200,300})
print(res)
```

### 7.intersection():

```python
s={1,2,3,4,5,300,400,100}
res=s.intersection({100,200,300})
print(res)
```

### 8.difference():

```python
s={1,2,3,4,5,300,400,100}
res=s.difference({100,200,300})
print(res)
```

### 9.symmetric_difference():

```python
s={1,2,3,4,5,300,400,100}
res=s.symmetric_difference({100,200,300})
print(res)
```
