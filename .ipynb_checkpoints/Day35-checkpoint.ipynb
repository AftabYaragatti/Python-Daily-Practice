# List built-in-methods:

### 1.append():

```python
l=[10,20,30,40,50]
l.append(60)
print(l)
```

```python
l=[10,20,30,40,50]
l.append([60,70,80,90,100])
print(l)
```

### 2.extend():

```python
l=[10,20,30,40,50]
l.extend([60,70,80,90,100])
print(l)
```

### 3.insert():

```python
l=[10,20,30,40,50]
l.insert(3,100)
print(l)
```

### 4.count():

```python
l=[10,20,30,40,50,30,30]
print(l.count(30))
```

### 5.index():

```python
l=[10,20,30,40,50]
print(l.index(30))
```

### 6.reverse():

```python
l=[10,20,30,40,50]
l.reverse()
print(l)
```

### 7.sort():

```python
l=[100,420,230,40,150]
l.sort()
print(l)
```

```python
l=[100,420,230,40,150]
l.sort(reverse=True)
print(l)
```

### 8.pop():

```python
l=[10,20,30,40,50]
l.pop()
print(l)
```

### 9.remove():

```python
l=[10,20,30,40,50]
l.remove(30)
print(l)
```

### 10.clear():

```python
l=[10,20,30,40,50]
l.clear()
print(l)
```
