# SCOPE OF VARIABLE

## 1. GLOBAL VARIABLE :

```python
## Access the varibale inside and outside the function
num=98
def display():
    print(f'The Number is Accesible Inside Function: {num}')
display()
print(f'The Number is Accesible Outside Function: {num}')
```

```python
## Global Key Word
num=98
def outer():
    global num
    print(f'Before Update Value is : {num}')
    num=num+7
    print(f'After Update Value is : {num}')
    num=90
    print(f'After Updating 2nd time Value is : {num}')
outer()
```

## 2. LOCAL VARIABLE :

```python
## Access the variable(which is present inside) inside function but not accesible by outside function
def display():
    num=98
    print(f'The Number is Accesible Inside Function (Local Variable): {num}')
display()
print(f'The Number is Accesible Outside Function (To Access Local variables outside the function we need to use nonlocal keyword :')
```

```python
## Access the variable(which is present inside) inside function but not accesible by outside function
def Outer():
    num=98
    print(f'The Number is Accesible Inside Outer Function (Local Variable): {num}')
    def inner():
        print('To Access outside function variable to inside we need to use nonlocal keyword : ')
        nonlocal num
        print(f'The Number is Accesible Inside Inner Function (Local Variable): {num}')
    inner()
Outer()
print(f'The Number is Accesible Outside Function (To Access Local variables outside the function we need to use nonlocal keyword :{num}')
```

# MATCH CASE STATEMENT IN PYTHON :

```python
## WAP to check day based on number
day=eval(input("Enter the Number between 1 to 7 to check day"))
match day:
    case 1:
        print('Monday')
    case 2:
        print('Tuesday')
    case 3:
        print('Wednesday')
    case 4:
        print('Thursday')
    case 5:
        print('Friday')
    case 6:
        print('Saturday')
    case 7:
        print('Sunday')
    case _:
        print('Invalide')
```

```python
## WAP to Perform Arithmetic Operations 
print('==========Welcome to Digital Calculator==========')
while True:
    num1=int(input('Enter the First Number :'))
    num2=int(input('Enter the Second Number :'))
    print('1.Addition\n2.Substraction\3.Multiplication\n4.Division\n5.Floor Division\n6.Power\n7.Modulas')
    num=int(input('Enter the Number to Perform Operation :'))
    match num:
        case 1:
            print(f'Sum : {num1+num2}')
        case 2:
            print(f'Sub : {num1-num2}')
        case 3:
            print(f'Mul : {num1*num2}')
        case 4:
            print(f'Div : {num1/num2}')
        case 5:
            print(f'Floor Div : {num1//num2}')
        case 6:
            print(f'Power : {num1**2}')
        case 7:
            print(f'Sum : {num1%num2}')
        case _:
            print(f'Invalid Choice')
    choice=input('Do You want to exit : (yes/no)')
    if choice=='Yes':
        break
```
