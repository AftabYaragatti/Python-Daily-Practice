## FOR LOOPING STATEMENS

#### SERIES PROGRAM

#### 1. WAP to print series of numbers from m to n inclusive by using for loop

```python
m=int(input())
n=int(input())
for i in range(m,n+1):
    print(i)
```

#### 2. WAP to print the sum of the numbers from m to n by using for loop

```python
m=int(input())
n=int(input())
sum=0
for i in range(m,n+1):
    sum+=i
print(sum)
```

#### 3. WAP to check the given number is prime or not

```python
num=int(input())
count=0
for i in range(1,num+1):
    if num%i==0:
        count+=1
if count==2:
    print('Prime Number')
else:
    print('Not a Prime Number')
```

#### OR

```python
num=int(input())
count=0
for i in range(2,num//2+1):
    if num%i==0: 
        count+=1
if count==0:
    print('Prime Number')
else:
    print('Not a Prime Number')
```

#### OR

```python
num=int(input())
count=0
for i in range(2,num//2+1):
    if num%i==0: 
        count+=1
        break
if count==0:
    print('Prime Number')
else:
    print('Not a Prime Number')
```

#### 4. WAP to check the given character is Vowel or not

```python
ch=input()
for i in 'aeiouAEIOU':
    if ch==i:
        print('Character is Vowel')
```

### PRINT CHARACTERS (STRING PROGRAMING)

#### 5. WAP to print the Upper case character from a given String

```python
st=input()
for ch in st:
    if ch>='A' and ch<='Z':
        print(ch)
```

#### 6. WAP to print the all vowels from a given string

```python
st=input()
for ch in st:
    if ch in 'aeiouAEIOU':
        print(ch)
```

#### 7. WAP to print the Symbols from a given string

```python
st=input()
for ch in st:
    if not((ch>='A' and ch<='Z') or (ch>='a' and ch<='z') or (ch>='0' and ch<='9')):
        print(ch)
```

#### 8. WAP to print the all digits from a given string

```python
st=input()
for ch in st:
    if (ch>='0' and ch<='9'):
        print(ch)
```
