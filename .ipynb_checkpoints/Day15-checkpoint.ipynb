## WHILE LOOPING STATEMWNTS (COUNTING/NUMBER):

#### 11. WAP to check given number is prime or not

```python
start=1  #initially take the start value as 1 
num=int(input()) #take the number from user, for example i will consider num value as 11
count=0 #take the count is zero always, becuase counter starts from 0 itself
while start<=num: # check the condition whether start value is less than or eual to number or not, if true enter to while loop
    if num%start==0: # check that number is divible by start or not, if divisible then it will enter to the if TSB
        count+=1 # update the count from 0 to 1
    start+=1  # increamennt the start value by 1 so in order to stop at certain condition
if count==2: # check count is equals to 2 or not, if it equals to 2 then it will enter to the if TSB
    print('Prime Number') # print prime number if condition is true
else: #if count is not equals to 2, if it not equals to 2 then it will enter to the else TSB
    print('Not a Prime Number') # print not prime number if condition is False
```

#### 12. WAP to print sum of the numbers from 1 to 5

```python
start=1 #initially take the start value as 1 
Sum=0 #take initially sum=0 always, becuase to do the sum or any number that starts from 0 itself
while start<=5: # check the condition whether start value is less than or eual to 5 or not, if true enter to while loop
    Sum+=start #update the sum value by adding sum+start
    start+=1 # update the start value by 1
print(Sum) # prnt the total sum of the given number
```

#### 13. WAP to print product of the numbers from 1 to n (Factorial of Number Program)

```python
start=1 #initially take the start value as 1 
n=int(input()) # take input from user, for example i will take 5 
prod=1 #take initially prod=1 always, becuase to do the product operations of any number that starts from 1 itself, anything with zero is always 0
while start<=n: # check the condition whether start value is less than or eual to n or not, if true enter to while loop
    prod*=start #update the prod value by multiplying prod*start
    start+=1 # update the start value by 1
print(prod) # prnt the total sum of the given number
```

#### 14. WAP to check sum of the divisors of a given number is equals to same number or not. If it is then print the Perfect Numbers, Else not a Perfect Number. Note : Do not conside the same number as a divisors, else you will not get result

```python
start=1 #initially take the start value as 1 
num=int(input()) # take input from user, for example i will take 6
Sum=0 #take initially sum=0 always, becuase to do the sum or any number that starts from 0 itself
while start<=num-1: # check the condition whether start value is less than or eual to num-1 or not, if true enter to while loop
    if num%start==0: # check that number is divible by start or not, if divisible then it will enter to the if TSB
        Sum+=start #update the sum value by adding sum+start
    start+=1 # update the start value by 1
print('Sum :',Sum) # print the total sum
if num==Sum: # check that num is equals to sum then it will enter to if TSB
    print('Perfect Number') #Print the Perfect Number
else:  # if num is not  equals to sum then it will enter to Else TSB
    print('Not a Perfect Number') #Print the Not a Perfect Number
```

#### 15. WAP to print the common divisors for 8 and 12 from 1 to 9

#### 16. WAP to print the GCD(Greatest Comman Divisor) of the given two numbers. or HCF(Highest Common Factor)

```python
n1=int(input()) #take the input from user , For example i will take as 8
n2=int(input()) #take the input from user , For example i will take as 12
start=1  #initially take the start value as 1 
divisors=0 #initially take the divisors value as 0
while start<=n1 and start<=n2: # check the condition whether start value is less than or eual to n1 and n2 or not, if true enter to while loop
    if n1%start==0 and n2%start==0: # check that number n1 is divible by start or not and number n2 is divible by start or not, 
                                   # if divisible then it will enter to the if TSB
        divisors=start #store the start value at divisors
    start+=1 #update the start value by 1
print(divisors) #print the divisors
```

#### 17. WAP to print the GCD(Greatest Comman Divisor) of the given three numbers. or HCF(Highest Common Factor)

```python
n1=int(input()) #take the input from user , For example i will take as 8
n2=int(input()) #take the input from user , For example i will take as 12
n3=int(input()) #take the input from user , For example i will take as 24
start=1  #initially take the start value as 1 
divisors=0 #initially take the divisors value as 0
while start<=n1 and start<=n2 and start<=n3: # check the condition whether start value is less than or eual to n1 and n2 and n3 or not, 
                                             # if true enter to while loop
    if n1%start==0 and n2%start==0 and n3%start==0: # check that number n1 is divible by start or not and number n2 is divible by start or not,
                                                     # and n3 also , if divisible then it will enter to the if TSB
        divisors=start #store the start value at divisors
    start+=1 #update the start value by 1
print(divisors) #print the divisors
```

#### 18. WAP to print the LCM(Least Common Multiple) of the given two numbers.

```python
n1=int(input()) #take the input from user , For example i will take as 8
n2=int(input()) #take the input from user , For example i will take as 12
start=1  #initially take the start value as 1 
gcd=0 #initially take the divisors value as 0
while start<=n1 and start<=n2: # check the condition whether start value is less than or eual to n1 and n2 or not, if true enter to while loop
    if n1%start==0 and n2%start==0: # check that number n1 is divible by start or not and number n2 is divible by start or not, 
                                   # if divisible then it will enter to the if TSB
        gcd=start #store the start value at divisors
    start+=1 #update the start value by 1
print(n1*n2//gcd) #print the divisors
```

#### 19. WAP to check sum of the numbers from 1 to n is prime number or not.

```python
start=1 #initially take the start value as 1
n=int(input()) #take the input from user , For example i will take as 7
Sum=0 #take initially sum=0 always, becuase to do the sum or any number that starts from 0 itself
count=0 #take the count is zero always, becuase counter starts from 0 itself
while start<=n: # check the condition whether start value is less than or eual to 5 or not, if true enter to while loop
    Sum+=start #update the sum value by adding sum+start
    start+=1 # update the start value by 1
print('Sum',Sum) # prnt the total sum of the given number
start=1 #initially take the start value as 1
while start<=Sum: # check the condition whether start value is less than or eual to number or not, if true enter to while loop
    if Sum%start==0: # check that number is divible by start or not, if divisible then it will enter to the if TSB
        count+=1 # update the count from 0 to 1
    start+=1  # increamennt the start value by 1 so in order to stop at certain condition
if count==2: # check count is equals to 2 or not, if it equals to 2 then it will enter to the if TSB
    print('Prime Number') # print prime number if condition is true
else: #if count is not equals to 2, if it not equals to 2 then it will enter to the else TSB
    print('Not a Prime Number') # print not prime number if condition is False
```
