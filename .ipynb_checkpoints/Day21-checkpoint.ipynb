## FOR LOOPING STATEMENS (Storing based program)

#### 15.WAP to reverse the string

```python
st=input()
res=''
for ch in st:
   res+=ch
print(res[::-1])
```

### OR (Without Slicing)

```python
st=input()
res=''
for ch in st:
   res=ch+res
print(res)
```

#### 16. WAP to store all the alphabets from a given string

```python
st=input()
res=''
for ch in st:
   if (ch>='A' and ch<='Z') or (ch>='a' and ch<='z'):
       res+=ch
print(res)
```

#### 17. WAP to convert the the uppercase to lowercase and consonant remains same, vowels should change to * and e should change to $

```python
st=input()
res=''
for ch in st:
   if (ch>='A' and ch<='Z') or (ch>='a' and ch<='z'):
       if ch>='A' and ch<='Z':
           ch=chr(ord(ch)+32)
           if ch in 'aeiouAEIOU':
               if ch=='E' or ch=='e':
                   res+='$'
               else:
                   res+='*'
           else:
               res+=ch
       else:           
           res+=ch
   else:
       res+=''
print(res)
```

#### 17. WAP to print the given string '123HELLO WORLD456', remove vowels and returns the sum of digits
#### Output : hlwrld21

```python
st=input()
res=''
sum=0
for ch in st:
    if (ch>='A' and ch<='Z') and ch not in 'AEIOU':
        res+=chr(ord(ch)+32)
    elif ch>='0' and ch<='9':
        sum+=int(ch)
print(res+str(sum))
```

#### 18. WAP to print the given string '123HELLO WORLD456', reverse the digits and convert uppercase consonant to lower case
#### Output : 321hEllO wOrld654

```python
st='123 HELLO WORLD 456'
digit=''
res=''
for i in st:
    if i>='0' and i<='9':
        digit+=i
    else:
        if digit!=' ':
            res+=digit[::-1]
            digit=''
        if (i>='A' and i<='Z') and i not in 'aeiouAEIOU':
            if i>='A' and i<='Z':
                res+=chr(ord(i)+32)
            else:
                res+=i
        else:
            res+=i
if digit==' ':
    res+=digit[::-1]
else:
    if digit!=' ':
            res+=digit[::-1]
            digit=''
print(res)
```

#### 21. Merging two string character by character 
#### I/P : S1:abc, S2:xyz
#### O/P : axbycz
#### I/P : S1:abcde, S2:xyz
#### O/P : axbyczde

```python
st1=input()
st2=input()
res=''
i=0
while i<len(st1) or i<len(st2):
    if i<len(st1) and i<len(st2):
        res+=st1[i]+st2[i]
    elif i<len(st1):
        res+=st1[i]
    else:
        res+=st2[i]
    i+=1
print(res)
```
