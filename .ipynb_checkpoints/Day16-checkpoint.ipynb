## WHILE LOOP WITH NUMBERS PROGRAMS :

#### 1. WAP to print all the digits from a given number

```python
num=int(input())
while num!=0:
    ld=num%10
    print(ld)
    num=num//10
```

#### 2. WAP to print the digits which are greater than 3 from a given number

```python
num=int(input())
start=1
while num!=0:
    ld=num%10
    if ld>3:
        print(ld)
    num=num//10
```

#### 3. WAP to print all odd digits in a given number

```python
num=int(input())
start=1
while num!=0:
    ld=num%10
    if ld%2!=0:
        print(ld)
    num=num//10
```

#### 4. WAP to print the sum of the digits in the given numbers

```python
num=int(input())
start=1
sum=0
while num!=0:
    ld=num%10
    sum+=ld
    num=num//10
print(sum)
```

#### 5. WAP to print the greatest digit from a given number

```python
num=int(input())
gd=num%10
while num!=0:
    ld=num%10
    if ld>gd:
        gd=ld
    num=num//10
print(gd)
```

#### 6. WAP to print the difference between greatest digit and smallest digit

```python
num=int(input())
gd=num%10
sd=num%10
while num!=0:
    ld=num%10
    if ld>gd:
        gd=ld
    if ld<sd:
        sd=ld
    num=num//10
print(gd-sd)
```

#### 7. WAP to check whether sum of digits and product of the digits in a given number is equal or not.If it is equals then print **SPY NUMBER** Else Not a SP NUMBER

### **SPY NUMBER PROGRAM**
#### Take any digits of a number, the sum of the digits are equals to the products of digits
#### **sum** of digits == **product** of digits
#### Example : 123
#### 1+2+3==1*2*3
#### 6==6

```python
num=int(input())
sum=0
prod=1
while num!=0:
    ld=num%10
    sum+=ld
    prod*=ld
    num=num//10
if sum==prod:
    print("SPY NUMBER")
else:
    print("NOT A SPY NUMBER")
```

#### 8. WAP to check sum of the digits in square of the given number is equals to the same number itself for not. If it is true then print **NEON NUMBER**

### **NEON NUMBER PROGRAM**
#### Take any number and that number should be square and for that **Squared Number** take each digit sum of Squared Number, that should equals to the original number
#### **Square** of the **Number** and do the sum of each **Square Numbers of digit** == **Original** Number
#### Example : num=9
#### sqr=9**2=81
#### 81=> 8 + 1 = 9
#### Original Number == Ssum of digit of Square number
#### 9 == 9

```python
num=int(input())
square=num**2
sum=0
while square!=0:
    ld=square%10
    sum+=ld
    square//=10
if sum==num:
    print("NEON Number")
else:
    print("Not A NEON Number")
```

#### 9. WAP to print the series of the numbers from 1 to each digit in a given number.

```python
num=int(input())
while num!=0:
    ld=num%10
    start=1
    while start<=ld:
        print(f'For digit {ld}, The series starts from 1 to {ld} : {start}')
        start+=1
    num=num//10
    print()
```
