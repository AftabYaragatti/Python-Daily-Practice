#### 11. WAP to print the final sum of the factorial of each digit in a given number (Fact - Product of the series from 1 to n)

### **STRONG NUMBER PROGRAM**
#### Take any number of the digits that should be the **sum** of the **factorial** of the number is equals to the original number
#### **Sum** of the **factorial(Product)** of the number == **Original** Number
#### Example : num=145
#### fact of each digit : 5 - 1*2*3*4*5=120, 4 - 1*2*3*4=24, 1 - 1 = 1
#### sum of fact : 120+24+1=145
#### Original Number == **Sum** of the **factorial(Product)** of the number
#### 145 == 145

```python
num=int(input())
num1=num
sum=0
while num!=0:
    ld=num%10
    start=1
    prod=1
    while start<=ld:
        prod*=start
        start+=1
    num//=10
    sum+=prod
    print(f'For digit {ld}, The Product of series starts from 1 to {ld} : {prod}')
print(f'Sum of all Product is {sum}')
if num1==sum:
    print('Strong Number')
else:
    print('Not a Strong Number')
```

#### 12. WAP to print count of the digits in the given number.

#### 13. WAP to count the Odd Digits from given number

```python
num=int(input())
count=0
while num!=0:
    ld=num%10
    if ld%2!=0:
        count+=1
    num//=10
print(count)
```

#### 14. WAP to power each digit with the count of the digits in the given number and print the power.

```python
num=int(input())
count=0
num1=num
while num!=0:
    count+=1
    num//=10
print('Count :',count)
while num1!=0:
    ld=num1%10
    print(ld**count)
    num1//=10
```

#### 15. WAP to check sum of all power of the digitws with its count of the total digits in the given number. **ARMSTRONG Number**

### **ARMSTRONG NUMBER PROGRAM**
#### Take any Number, the length(count) of the number should be powerd by each digit of the number and do the sum for each powered digit, that should eqauls to the original number.
#### Sum of the each Power digit of the given number length == Original Number
#### Example : num = 153
#### len/count = 153 -> 3
#### 1**3 + 5**3 + 3**3 = 1 + 125 + 27 = 153
#### Original Number == Sum of the each Power digit of the given number length
#### 153 == 153

```python
num=int(input())
count=0
num1=num
while num!=0:
    count+=1
    num//=10
print('Count :',count)
Sum=0
while num1!=0:
    ld=num1%10
    num1//=10
    Sum=Sum+(ld**count)
print(Sum)
if num1==Sum:
    print('Armstrong Number')
else:
    print('Not a Armstrong Number')
```
