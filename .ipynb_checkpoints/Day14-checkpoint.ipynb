## WHILE LOOPING STATEMWNTS (SERIES LOGIC):

#### 1. Write a Program to print Series of numbers from 1 to 10.

```python
num=int(input()) #take the input from user as 10 given in the questions
start=1 # initially take the start=1 becuase we need to generate the series from 1 to 10
while start<=num: # in while condition we need to generate the series from 1 to 10 and Check the condition whether number is greater than 0 or not , 
                  # if it is true then it will enter to the whike loop 
    print(start) #print the number every time when the condition true , print the number,where the number is printing on different data 
                 # on each iteration value will be chnange
    start+=1 # we need to increment the start value because to get the number sseries one after and to stop the while loop by increment by 1
```

#### 2. WAP to print series from 5 to 1

```python
start=5 # initially take the start=5 becuase we need to generate the series from 5 to 1 in decending order
while start>0:  # in while condition we need to generate the series from 5 to 1 and Check the condition whether number is greater than 0 or not , 
                # if it is true then it will enter to the whike loop 
    print(start) #print the number every time when the condition true , print the number,where the number is printing on different data 
                 # on each iteration value will be chnange
    start-=1 # we need to increment the start value because to get the number sseries one after and to stop the while loop by decrement by 1
```

#### 3. WAP to print series of alphabets from Capital A to E

```python
ch='A' #initially take the character from "A"
while ch<='E': # Check the condition whether character is less than or equal to E or not
    print(ch) # print character if the while condition is true
    ch=chr(ord(ch)+1) # take the ascii value of ch then increment by 1 then we will get next asciee value of character 
                       # that ascii value of character should convert to character and update the ch
```

#### 4. WAP to print the numbers which are divisble by 3 from 1 to 7

```python
start=1  #initially take the start value as 1  
while start<=7: # check the condition whether start value is less than or eual to 7 or not, if true enter to while loop
    if start%3==0: # check that start number is divisible by 3 or not, if divisible then it will enter to the if TSB
        print(start) # print if the above condtion true
    start+=1 # increamennt the start value by 1 so in order to stop at certain conditions
```

#### 5. WAP to print the numbers which are divisible by both 3 and 5 from 1 to 15

```python
start=1  #initially take the start value as 1  
while start<=15: # check the condition whether start value is less than or eual to 15 or not, if true enter to while loop
    if start%3==0 and start%5==0: # check that start number is divisible by 3 and 5 both or not, if divisible then it will enter to the if TSB
        print(start) # print if the above condtion true
    start+=1 # increamennt the start value by 1 so in order to stop at certain conditions
```

#### 6. WAP to print divisors of 6 from 1 to 5

```python
start=1  #initially take the start value as 1  
num=6  #initially take the number value as 6   
while start<=5: # check the condition whether start value is less than or eual to 5 or not, if true enter to while loop
    if num%start==0: # check that number is divible by start or not, if divisible then it will enter to the if TSB
        print(start) # print if the above condtion true
    start+=1  # increamennt the start value by 1 so in order to stop at certain conditions
```

#### 7. WAP to print divisor of a given number

```python
start=1  #initially take the start value as 1  
num=int(input())  #take the number from user, for example i will consider num value as 9  
while start<=num: # check the condition whether start value is less than or eual to number or not, if true enter to while loop
    if num%start==0: # check that number is divible by start or not, if divisible then it will enter to the if TSB
        print(start) # print if the above condtion true
    start+=1  # increamennt the start value by 1 so in order to stop at certain conditions
```

#### 8. WAP to print Odd divisors of a given number

```python
start=1  #initially take the start value as 1  
num=int(input())  #take the number from user, for example i will consider num value as 18 
while start<=num: # check the condition whether start value is less than or eual to number or not, if true enter to while loop
    if start%2!=0: # check that start value is not divisible by 2 or not, if not divisible then it will enter to the if TSB
        if num%start==0: # check that number is divible by start or not, if divisible then it will enter to the if TSB
            print(start) # print if the above condtion true
    start+=1  # increamennt the start value by 1 so in order to stop at certain conditions
```

#### 9. WAP to count and print the total numbers which are divisible by 4, from 1 to 9

```python
start=1  #initially take the start value as 1  
count=0  #take the count is zero always, becuase counter starts from 0 itself 
while start<=9: # check the condition whether start value is less than or eual to 9 or not, if true enter to while loop
    if start%4==0: # check that start value is divisible by 4 or not, if divisible then it will enter to the if TSB
        count+=1 # update the count from 0 to 1
    start+=1  # increamennt the start value by 1 so in order to stop at certain condition
print(count) #print the total count from 1 to 9
```

#### 10. WAP to check the count of the divisors of a given number is equals to 2 or not. If it is equals then print the prime number, else print not a prime number

```python
start=1  #initially take the start value as 1 
num=int(input()) #take the number from user, for example i will consider num value as 11
count=0 #take the count is zero always, becuase counter starts from 0 itself
while start<=num: # check the condition whether start value is less than or eual to number or not, if true enter to while loop
    if num%start==0: # check that number is divible by start or not, if divisible then it will enter to the if TSB
        count+=1 # update the count from 0 to 1
    start+=1  # increamennt the start value by 1 so in order to stop at certain condition
print('Count :',count) # print the count 
if count==2: # check count is equals to 2 or not, if it equals to 2 then it will enter to the if TSB
    print('Prime Number') # print prime number if condition is true
else: #if count is not equals to 2, if it not equals to 2 then it will enter to the else TSB
    print('Not a Prime Number') # print not prime number if condition is False
```
